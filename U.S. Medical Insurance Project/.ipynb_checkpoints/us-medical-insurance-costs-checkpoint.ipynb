First it is important to get the data from the file given __insurance.csv__ 

For this the *csv* module is needed to get the data from the file given. And Throught this module we can get all the information into a list of dictionaries that contain ech data of the patients.

```python
import csv
patients_list = []
with open("insurance.csv") as insurances:
    insurances_reader = csv.DictReader(insurances)
    for insurance in insurances_reader:
        patients_list.append(insurance)
        #print(insurance)

print(patients_list[1])
```

Now the data is stored in a list of ditionaries, There are multiple ways to work with this data. One way is creating a class called __Patient__ storing each patients data thre an work with them as we want creating different methods.

```python
class Patient:
    patient_id = 0
    def __init__(self, patient):
        # checking if the info to initiate the class is correct
        if type(patient) != dict:
            print("Error: patient given isnt a dictionary type")
            return None
        elif len(patient) != 7:
            print("Patient info incomplete or more info than expected")
            return None
        # initiating the class
        self.age = patient["age"]
        self.sex = patient["sex"]
        self.bmi = patient["bmi"]
        self.children = patient["children"]
        self.smoker = patient["smoker"]
        self.region = patient["region"]
        self.charges = patient["charges"]
        # There can be added info if needed, lets make sure to change the the parameters in the exceptions handler
    def __repr__(self):
       return """ Patients Info:
        AGE: {age}
        SEX: {sex}
        BMI: {bmi}
        Number of children: {children}
        Smoker: {smoker}
        Region: {region}
        Insurance Costs: [charges]""".format(age = self.age, sex = self.sex, bmi = self.bmi, children = self.children, smoker = self.smoker, region = self.region, charges = self.charges)
 
    #If we want to convert our class back to a dictionary , lets use this function
    def patient_to_dictionary(self):
        patient_dict ={}
        patient_dict["age"] = self.age
        patient_dict["sex"] = self.sex
        patient_dict["bmi"] = self.bmi
        patient_dict["children"] = self.children
        patient_dict["smoker"] = self.smoker
        patient_dict["region"] = self.region
        patient_dict["charges"] = self.charges
        return patient_dict
    
    # THis next methods return the data asked for from the Patient class
    def get_patient_age(self):
        return self.age
        
    def get_patient_sex(self):
        return self.sex
    
    def get_patient_bmi(self):
        return self.bmi
    
    def get_patient_children(self):
        return self.children
    
    def get_patient_smoker(self):
        return self.smoker
    
    def get_patient_region(self):
        return self.region
    
    def get_patient_charges(self):
        return self.charges
    
    
```

The class is ready to be used. Next step is making the list that has all the patiets info saved in dictionaries into a list of the class __Patient__.

```python
patients_list_class = []
for patient in patients_list:
    patients_list_class.append(Patient(patient))
    
print(patients_list_class[0])
```

Now we have a way to identify our patiens and dont lose their order from the data tha was given to us. Now lets separate our patients by region 

```python
def patients_by_region_order(patients):
    patients_by_region_dict = {"northeast": [], "northwest": [], "southwest": [], "southeast": []}
    for patient in patients:
        if patient.region == "northeast":
            patients_by_region_dict["northeast"].append(patient)
        elif patient.region == "northwest":
            patients_by_region_dict["northwest"].append(patient)
        elif patient.region == "southeast":
            patients_by_region_dict["southeast"].append(patient)
        elif patient.region == "southwest":
            patients_by_region_dict["southwest"].append(patient)
    return patients_by_region_dict

patients_by_region = patients_by_region_order(patients_list_class)
print(patients_by_region)
```
